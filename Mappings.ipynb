```python
h_set = set({})
for i in hlts.unique():
    for j in i[1:-1].split(','):
        h_set.add(j.strip()[1:-1])
```

<h3>Getting Rating Based On Data Types</h3>

```python
mappings = pd.DataFrame(columns=["establishment","highlights","count","mean","std","min","max","25%","50%","75%"])
mappings.head()
```

```python
def filtering1(establishment, highlight):
    return dataset[(dataset["establishment"]==establishment) & (dataset["highlights"].apply(lambda x : highlight in x))][["aggregate_rating"]].describe()
```

```python
i = "Quick Bites"
j = "Lunch"
res = filtering1(i,j)["aggregate_rating"]
print(res)
mappings.append({"establishment":i,"highlights":j,"count":res["count"],"mean":res["mean"],"std":res["std"],"min":res["min"],"max":res["max"],"25%":res["25%"],"50%":res["50%"],"75%":res["75%"]}, ignore_index=True)
```

```python
dataset["timings"]
```

```python
def func3(ts):
    arr = list(filter(lambda x : int(x)<25, re.findall("\d+",ts)))
    l = len(arr)
    if l < 2 :
        return 0
    fh =  int(arr[0])
    if(fh<5):
        fh += 12
    return fh
def func4(ts):
    arr = list(filter(lambda x : int(x)<25, re.findall("\d+",ts)))
    l = len(arr)
    if l < 2 :
        return 24
    sh = int(arr[1])
    if(sh>2 or sh!=12):
        sh += 12
    if(sh==24):
        sh = 0
    return sh
```

```python
dataset["open"] = dataset["timings"].apply(func3)
dataset["close"] = dataset["timings"].apply(func4)
```

```python
dataset.drop(["timings"], inplace=True, axis=1)
dataset.head(3)
```

```python
dataset.to_csv("time_corrected.csv", index=False)
```

<h3> Including the time factor </h3>

```python
mappings2 = pd.DataFrame(columns=["hour","establishment","highlights","count","mean","std","min","max","25%","50%","75%"])
mappings2.head()
```

```python
def filtering2(t,establishment, highlight):
    return dataset[((dataset["open"]>t)&(dataset["close"]<t))|((dataset["open"]>t)&(dataset["close"]>t)&(dataset["close"]<2)) & (dataset["establishment"]==establishment) & (dataset["highlights"].apply(lambda x : highlight in x))][["aggregate_rating"]].describe()
```

```python
i = "Quick Bites"
j = "Lunch"
h = 11
res = filtering2(h,i,j)["aggregate_rating"]
print(res)
mappings2.append({"hour":h,"establishment":i,"highlights":j,"count":res["count"],"mean":res["mean"],"std":res["std"],"min":res["min"],"max":res["max"],"25%":res["25%"],"50%":res["50%"],"75%":res["75%"]}, ignore_index=True)
```

```python
for i in est_list :
    for j in list(h_set):
        for h in range(24):
            res = filtering2(h,i,j)["aggregate_rating"]
            mappings2 = mappings2.append({"hour":h,"establishment":i,"highlights":j,"count":res["count"],"mean":res["mean"],"std":res["std"],"min":res["min"],"max":res["max"],"25%":res["25%"],"50%":res["50%"],"75%":res["75%"]}, ignore_index=True)

mappings2.head()
```
